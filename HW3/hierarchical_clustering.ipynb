```python
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('./data/EastWestAirlines.csv')
scaler = MinMaxScaler(feature_range=(0, 1))
data_norm = scaler.fit_transform(data)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
Z = linkage(data_norm, method='complete')

dendrogram(Z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluserting = AgglomerativeClustering(n_clusters=2)

labels = cluserting.fit_predict(data_norm)
print(f'{labels=}') 
```

```python
from sklearn.decomposition import PCA
import numpy as np

pca = PCA(n_components=0.95)
pca.fit(data_norm)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python

pca = PCA(n_components=2)
X_pca = pca.fit_transform(data_norm)
plt.figure(figsize=(10, 7))
Z = linkage(X_pca, method='complete')
dendrogram(Z)
plt.show()
```
